```python
import tkinter as tk
from tkinter import scrolledtext

class VentanaPruebaExcepciones(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Prueba de Excepciones - POO")
        self.geometry("500x400")
        self.resizable(False, False)
        self.config(bg="#f4f4f4")

        # ===== Área de texto con scroll =====
        self.area_texto = scrolledtext.ScrolledText(
            self, 
            wrap=tk.WORD, 
            font=("Consolas", 12), 
            height=15
        )
        self.area_texto.pack(fill=tk.BOTH, expand=True, padx=10, pady=(10, 5))
        self.area_texto.config(state=tk.NORMAL)

        # ===== Panel de botones (siempre visible abajo) =====
        frame_botones = tk.Frame(self, bg="#f4f4f4")
        frame_botones.pack(side=tk.BOTTOM, pady=10)

        btn_bloque1 = tk.Button(
            frame_botones, text="Ejecutar Bloque 1",
            command=self.ejecutar_bloque1,
            bg="#4CAF50", fg="white",
            font=("Arial", 10, "bold"), width=16
        )
        btn_bloque1.grid(row=0, column=0, padx=5)

        btn_bloque2 = tk.Button(
            frame_botones, text="Ejecutar Bloque 2",
            command=self.ejecutar_bloque2,
            bg="#2196F3", fg="white",
            font=("Arial", 10, "bold"), width=16
        )
        btn_bloque2.grid(row=0, column=1, padx=5)

        btn_limpiar = tk.Button(
            frame_botones, text="Limpiar",
            command=self.limpiar,
            bg="#f44336", fg="white",
            font=("Arial", 10, "bold"), width=12
        )
        btn_limpiar.grid(row=0, column=2, padx=5)

    # ===== Bloque 1 =====
    def ejecutar_bloque1(self):
        self.area_texto.insert(tk.END, "Ingresando al primer try\n")
        try:
            cociente = 10000 / 0  # Lanza ZeroDivisionError
            self.area_texto.insert(tk.END, "Después de la división\n")
        except ZeroDivisionError:
            self.area_texto.insert(tk.END, "División por cero\n")
        finally:
            self.area_texto.insert(tk.END, "Ingresando al primer finally\n\n")

    # ===== Bloque 2 =====
    def ejecutar_bloque2(self):
        self.area_texto.insert(tk.END, "Ingresando al segundo try\n")
        try:
            objeto = None
            objeto.toString()  # En Python lanzará AttributeError
            self.area_texto.insert(tk.END, "Imprimiendo objeto\n")
        except ZeroDivisionError:
            self.area_texto.insert(tk.END, "División por cero\n")
        except Exception:
            self.area_texto.insert(tk.END, "Ocurrió una excepción\n")
        finally:
            self.area_texto.insert(tk.END, "Ingresando al segundo finally\n\n")

    # ===== Limpiar área =====
    def limpiar(self):
        self.area_texto.delete("1.0", tk.END)


# ===== MAIN =====
if __name__ == "__main__":
    app = VentanaPruebaExcepciones()
    app.mainloop()
```
